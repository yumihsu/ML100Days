```python
import os
import keras
import itertools
# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
def build_mlp1(input_shape , output_unit = 10 , num_nerous = [512,256,128,64]):
    input_layer = keras.layers.Input(input_shape)
    for i, num in enumerate(num_nerous):
        if i == 0 :
            x = keras.layers.Dense(units = num,
                                  activation = 'relu',
                                  name = 'hidden_layer'+str( i+1))(input_layer)
            x = keras.layers.BatchNormalization()(x)
        else :
            x = keras.layers.Dense(units = num ,
                                  activation = 'relu',
                                  name = 'hidden_layer'+str( i+1))(x)
            x = keras.layers.BatchNormalization()(x)
    output = keras.layers.Dense(units = output_unit,
                               activation = 'softmax',
                               name = 'output_layer')(x)
    model = keras.models.Model(inputs = [input_layer],
                              outputs = [output])        
    return model
model1 = build_mlp1(input_shape=x_train.shape[1:])
```

```python
def build_mlp2(input_shape , output_unit = 10 , num_nerous = [512,256,128,64]):
    input_layer = keras.layers.Input(input_shape)
    for i, num in enumerate(num_nerous):
        if i == 0 :
            x = keras.layers.BatchNormalization()(input_layer)
            x = keras.layers.Dense(units = num,
                                  activation = 'relu',
                                  name = 'hidden_layer'+str( i+1))(x)
            
        else :
            x = keras.layers.BatchNormalization()(x)
            x = keras.layers.Dense(units = num ,
                                  activation = 'relu',
                                  name = 'hidden_layer'+str( i+1))(x)
    output = keras.layers.Dense(units = output_unit,
                               activation = 'softmax',
                               name = 'output_layer')(x)
    model = keras.models.Model(inputs = [input_layer],
                              outputs = [output])        
    return model
model2 = build_mlp2(input_shape=x_train.shape[1:])
```

```python
"""Code Here
設定超參數
"""
EPOCH = 20
BATCH_SZIE = [256,128,32,16,2]
LEARNING_RATE = 1e-3

optimizers = keras.optimizers.Adam(lr=LEARNING_RATE)


model1.compile(loss='categorical_crossentropy',
            metrics= ['acc'],
             optimizer = optimizers)
model2.compile(loss='categorical_crossentropy',
            metrics= ['acc'],
             optimizer = optimizers)
```

```python
results = {}
for b in BATCH_SZIE:
    model1.fit (x_train,y_train, 
                epochs=EPOCH,
               batch_size= int(b),
               validation_data=(x_test,y_test),
               shuffle=True)
    results['model1_Batchsize_'+str(b)]={'train_loss' : model1.history.history["loss"],
                                        'valid_loss' : model1.history.history["val_loss"],
                                        'train_acc' : model1.history.history["acc"],
                                        'valid_acc' : model1.history.history["val_acc"]}
    model2.fit (x_train,y_train, 
            epochs=EPOCH,
           batch_size= int(b),
           validation_data=(x_test,y_test),
           shuffle=True)
    results['model2_Batchsize_'+str(b)]={'train_loss' : model2.history.history["loss"],
                                        'valid_loss' : model2.history.history["val_loss"],
                                        'train_acc' : model2.history.history["acc"],
                                        'valid_acc' : model2.history.history["val_acc"]}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
"""Code Here
將結果繪出
"""
color_bar = ['r','g','b','y','c','m','k','lime','pink','olive']
plt.figure(figsize=(16,16))
for i , cond in enumerate(results.keys()):

    plt.plot(range(len(results[cond]['train_loss'])),results[cond]['train_loss'],
             '-',color=color_bar[i],label = 'train_loss' + str(cond))
    plt.plot(range(len(results[cond]['valid_loss'])),results[cond]['valid_loss'],
             '--',color=color_bar[i],label = 'valid_loss' + str(cond))
    plt.title('loss')
    plt.legend(loc='best')
plt.ylim(0,3)
plt.show()


plt.figure(figsize=(16,16))
for i , cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train_acc'])),results[cond]['train_acc'],
             '-',color=color_bar[i],label = 'train_acc' + str(cond))
    plt.plot(range(len(results[cond]['valid_acc'])),results[cond]['valid_acc'],
             '--',color=color_bar[i],label = 'valid_acc' + str(cond))
    
    plt.title('acc')
    plt.legend(loc='best')

plt.show()
```
