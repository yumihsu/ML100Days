```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice

import warnings , matplotlib.cbook 

warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
# 執行各種樣板資料的繪圖迴圈
```

```python
# 設定群集與繪圖參數
plt.figure(figsize=(10, 8))
plot_num = 1
default_base = {'n_neighbors': 9, 'n_clusters': 3}
datasets =\
    [(noisy_moons, {'n_clusters': 2}), (blobs, {'n_clusters':4}), (no_structure, {'n_neighbors':3})]
for i_data,(dataset,data) in enumerate(datasets):
    params=default_base.copy()
    params.update(data)
    X,y=dataset
    X= StandardScaler().fit_transform(X)
    
    
    ward=cluster.AgglomerativeClustering(n_clusters=params['n_clusters'],linkage='ward')
    complete=cluster.AgglomerativeClustering(n_clusters=params['n_clusters'],linkage='complete')
    average=cluster.AgglomerativeClustering(n_clusters=params['n_clusters'],linkage='average')
    clustering_algorithms=[('ward',ward),
                           ('complete',complete),
                           ('average',average)]
    
    
    for name , algorithm in clustering_algorithms:
        if hasattr(algorithm,'labels_'):
            y_pred= algorithm.labels_.astype(np.int)
        else:
            y_pred = algorithm.fit_predict(X)
        
        plt.subplot(len(datasets),len(clustering_algorithms),plot_num)
        if i_data == 0 :
                    plt.title(name,size=14)
        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                     '#f781bf', '#a65628', '#984ea3',
                                     '#999999', '#e41a1c', '#dede00']),int(max(y_pred) + 1))))
        plt.scatter(X[:,0],X[:,1],s=10,color=colors[y_pred])
        
        plot_num += 1
```
