# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def relu(x):
    return abs(x) * (x > 0)
def dReLU(x):
    return (1 * (x > 0))
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, relu(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.title('Relu Function')
plt.ylim(-1,11)
plt.show()
```

```python
plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.title('Relu Function')
plt.ylim(-1,2)
plt.show()
```
